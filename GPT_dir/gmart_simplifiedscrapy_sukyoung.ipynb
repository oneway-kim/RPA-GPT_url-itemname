```python
best_links = []
source = requests.get('https://www.gmarket.co.kr/n/best').text
doc = SimplifiedDoc(source)
infos = doc.getElements('a', attr='class', value='itemname')
for info in infos:
    if info:
        link = info['href']
        product = info['html']
        best_links.append([product,link])
```

```python
df=pd.DataFrame(best_links)
df.columns=['product','link']
```

```python
df.to_excel('best_links.xlsx')
```

```python
i=0
product_infos_list={}
def product_infos(url):
    source = requests.get(url).text
    doc = SimplifiedDoc(source)
    product = doc.getElements('h1', attr='class', value='itemtit')
    price = doc.getElements('strong', attr='class', value='price_real')
    return product[0]['html'], price[0]['html']
```

```python
iter_list = df['link']
final_info= []
```
